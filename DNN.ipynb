```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
```

```python
#load the data 
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
#Normalize the data 
X_train = X_train / 255
X_test = X_test / 255
```

```python
#flatten the images because NN cannot process 2D images
X_train = X_train.reshape((-1,784))
X_test = X_test.reshape((-1,784))
```

- first layer - takes 784 inputs, has 128 neurons learns basic patterns like edges and stroke
- second layer - takes output from layer 1, combines edges into shapes
- third layer - learns high level features, understand digit structure
- output layer - 10 neurons because category is from 0 to 9

```python
# building the NN 
model = models.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',       #smart optimizer that adjusts learning speed
    loss='sparse_categorical_crossentropy', #measures prediction error
    metrics=['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs=5, batch_size=32)    #epochs means number of times the model will see the entire dataset
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test accuracy: {accuracy}')
```
